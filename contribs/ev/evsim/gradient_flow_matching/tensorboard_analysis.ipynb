```python
from tbparse import SummaryReader
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
sns.set_palette("muted")
```

```python
df_10 = SummaryReader("/home/isaacp/repos/EvMatsim/contribs/ev/evsim/gradient_flow_matching/utah_flow_results/0423134432_nclusters_10_utahnetwork/logs").scalars
df_50 = SummaryReader("/home/isaacp/repos/EvMatsim/contribs/ev/evsim/gradient_flow_matching/utah_flow_results/0423135446_nclusters_50_utahnetwork/logs").scalars
df_100 = SummaryReader("/home/isaacp/repos/EvMatsim/contribs/ev/evsim/gradient_flow_matching/utah_flow_results/0423140513_nclusters_100_utahnetwork/logs").scalars
df_200 = SummaryReader("/home/isaacp/repos/EvMatsim/contribs/ev/evsim/gradient_flow_matching/utah_flow_results/0423142014_nclusters_200_utahnetwork/logs").scalars
```

```python
df_10.head()
```

```python
df_10['tag'].unique()
```

```python
df_dict = {"10 Clusters":df_10, "50 clusters": df_50, "100 clusters":df_100, "200 clusters": df_200}
```

```python
for key, df in df_dict.items():
    mad_loss = df[df['tag'] == 'Logs/mad']
    plt.plot(mad_loss['step'], mad_loss['value'], label=key)

plt.legend()
plt.title("Mean Average Absolute Difference \n For Different Cluster Sizes")
plt.xlabel("Step")
plt.ylabel("MAAD")
plt.show()

for key, df in df_dict.items():
    mad_loss = df[df['tag'] == 'Loss/mse']
    plt.plot(mad_loss['step'], mad_loss['value'], label=key)
    
plt.legend()
plt.title("Mean Squared Error \n For Different Cluster Sizes")
plt.xlabel("Step")
plt.ylabel("MSE")
plt.show()
```
