```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 150)

y = (
    2e-5 * (x**2) - 0.0015 * x + 0.1331
)  # https://www.sciencedirect.com/science/article/pii/S1361920915301280

plt.plot(x, y)
plt.title("Energy Consumption vs Speed")
plt.xlabel("Speed (Km/h)")
plt.ylabel("Consumption (Kwh/Km)")
plt.show()
```

```python
"""linear regression coefficient predictor based on road gradient - https://www.sciencedirect.com/science/article/pii/S1361920917303887"""

x = np.linspace(-0.15, 0.15)

y = 81.159 * x**3 + 3.0407 * x**2 + 2.7996 * x + 0.0101

plt.plot(x, y)
plt.title("")
plt.xlabel("Road Gradient")
plt.ylabel("Coefficient (kwh/km)")
plt.show()
```

```python
gradient = -0.02
link_len_km = 100
link_len_km = np.log(link_len_km)
avg_speed_km_h = 90
avg_speed_km_h = np.log(avg_speed_km_h)

link_len_coeff = -0.019
avg_speed_coeff = -0.366


def grad_coeff_f(x):
    return 81.159 * x**3 + 3.0407 * x**2 + 2.7996 * x + 0.0101


grad_coeff = grad_coeff_f(gradient)

constant = 0.372


def consumption_f(x):
    return (
        constant
        + link_len_km * link_len_coeff
        + avg_speed_km_h * avg_speed_coeff
        + grad_coeff_f(x)
    )
```

```python
x = np.linspace(-0.15, 0.15)

y = consumption_f(x)

plt.plot(x, y)
plt.title("")
plt.xlabel("Road Gradient")
plt.ylabel("Consumption (kwh/km)")
plt.show()
```
